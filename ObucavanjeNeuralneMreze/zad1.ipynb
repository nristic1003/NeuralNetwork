```python
import numpy as np
import math
import matplotlib.pyplot as plt
from random import random
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from tensorflow.keras.optimizers import Adam
from sklearn.linear_model import LinearRegression 

A = 7
B = 5
f1 = 20
f2 = 6
P = 2
Q = 5
h = []
s = []
y = []
pi = math.pi
mean = 0;
indeksi = np.arange(0, 1, 0.0002)

for x in indeksi:
     h.append(A*math.sin(2*pi*f1*x)+B*math.sin(2*pi*f2*x))
     s.append(0.2*B*random())
     y.append(h[-1]+s[-1])

h = np.array(h)
s = np.array(s)
y = np.array(y)

plt.rcParams['figure.figsize'] = [8, 8]
plt.plot(indeksi , y, label='y(x)')
plt.plot(indeksi , h, label='h(x)')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)
```

```python
x_train, x_test, y_train, y_test = train_test_split(indeksi, y, test_size = 0.2)
```

```python

model = keras.Sequential()
model.add(keras.layers.Dense(32, input_shape = [1], activation = "relu"))
model.add(keras.layers.Dense(35, activation = "tanh"))
model.add(keras.layers.Dense(35, activation = "relu"))
model.add(keras.layers.Dense(45, activation = "tanh"))
model.add(keras.layers.Dense(45, activation = "relu"))
model.add(keras.layers.Dense(35, activation = "tanh"))
model.add(keras.layers.Dense(35, activation = "relu"))
model.add(keras.layers.Dense(20, activation = "tanh"))
model.add(keras.layers.Dense(32, activation = "relu"))

model.add(keras.layers.Dense(1))
opt = Adam(lr = 0.00107)
model.compile(optimizer = opt, loss='mse', metrics=['mae'])
history = model.fit(x_train, y_train, epochs = 3000, batch_size=900, validation_split = 0.15, verbose = 2)
```

```python
history_dict = history.history

loss_values = history_dict['loss']
val_loss_values=history_dict['val_loss']
plt.plot(loss_values,'b',label='training loss')
plt.plot(val_loss_values,'r',label='val training loss')
plt.legend()
plt.xlabel("Epochs")
loss_values = np.array(loss_values)
print(np.amin(loss_values))
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
print("The R2 score on the Train set is:\t{:0.3f}".format(r2_score(y_train, y_train_pred)))
print("The R2 score on the Test set is:\t{:0.3f}".format(r2_score(y_test, y_test_pred)))
```

```python
accuracy_values = history_dict['mae']
val_accuracy_values=history_dict['val_mae']
plt.plot(val_accuracy_values,'-r',label='val_mae')
plt.plot(accuracy_values,'-b',label='mae')
plt.legend()
```
